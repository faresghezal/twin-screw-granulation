```python
import numpy as np
from numpy import nan
import pandas as pd
df1=pd.read_excel("5,7 rpm vissza.xlsx", index_col=0)
df2=pd.read_excel("Gyártás.xlsx", index_col=0)
df3=pd.read_excel("5,7 rpm.xlsx", index_col=0)
df4=pd.read_excel("6,1 rpm vissza.xlsx", index_col=0)
df5=pd.read_excel("6,1 rpm.xlsx", index_col=0)
df6=pd.read_excel("6,5 rpm.xlsx", index_col=0)
df=pd.concat([df1, df2, df3, df4, df5, df6])

df
```

```python
df=df.drop(['name'], axis=1)
df=df.dropna()
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix,plot_confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import PolynomialFeatures, MinMaxScaler, StandardScaler
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neural_network import MLPClassifier
num_cols = list(df.dtypes[df.dtypes != 'object'].index.values) 
X = df[df.columns.difference(['type'])]
y = df['type']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
numerical = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler()),
    ('poly', PolynomialFeatures(3))])
column_preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical, num_cols)])
```

```python

clf1 = Pipeline(steps=[('preprocessor', column_preprocessor),
                      ('classifier', MLPClassifier(max_iter=1700,activation = 'logistic',solver='adam', random_state=1))])
clf2 = Pipeline(steps=[('preprocessor', column_preprocessor),
                      ('classifier',RandomForestClassifier(n_estimators=50, random_state=1))])
clf3 = KNeighborsClassifier()
eclf1 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='hard')
eclf1 = eclf1.fit(X, y)
y_pred = eclf1.predict(X_test)
print(f1_score(y_test, y_pred, average='micro'))
print(precision_score(y_test, y_pred, average="macro"))
print(recall_score(y_test, y_pred, average="macro"))
print(classification_report(y_test,y_pred))
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))

plot_confusion_matrix(eclf1, X_test, y_test, ax=ax)
```
